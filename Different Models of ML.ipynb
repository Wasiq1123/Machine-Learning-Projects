```python
x = sd1.drop(columns=sd.columns[81])
y = sd1.iloc[:,81]
print(y.shape)
print(y.name)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train1, y_test1 = train_test_split(x, y, test_size=0.25)
print(x_train.shape,y_train1.shape,x_test.shape,y_test1.shape)
y_train=y_train1.to_numpy().reshape(-1,1)
y_test=y_test1.to_numpy().reshape(-1,1)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
models = [
    LinearRegression(),
    Ridge(),
    Lasso(),
    RandomForestRegressor(),
    GradientBoostingRegressor()
]
```

```python
best_model = None
best_mse = float('inf')

for model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)

    print(f"{type(model).__name__} MSE: {mse}")
    if mse < best_mse:
        best_model = model
        best_mse = mse
```

```python
print(f"\nBest Model: {type(best_model).__name__} with MSE: {best_mse}")
```
