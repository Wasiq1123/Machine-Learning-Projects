```python
print(df[0:5])
a=['customerID','InternetService']
sd=df.drop(a,axis=1)
sd['MultipleLines'] = sd['MultipleLines'].replace("No phone service", 'No')
sd['gender']=sd['gender'].replace({'Female':0,'Male':1})
```

```python
s=sd['PaymentMethod'].value_counts()
print(s)
```

```python
print(sd['Contract'].value_counts())
print(sd[0:1])
```

```python
print(sd[0:5])
a=['Partner','Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for i in a:
  sd[i]=sd[i].replace({'Yes':1,'No':0})
```

```python
print(sd.head(3))
```

```python
columns_to_encode = ['Contract', 'PaymentMethod']
l = pd.get_dummies(sd[columns_to_encode])
sd1 = pd.concat([l, sd], axis=1)
print(sd1)
sd2=sd1.drop(columns_to_encode,axis=1)
```

```python
print(sd2.shape)
```

```python
print(sd2.head(3))
```

```python
print(sd2.info())
```

```python
g=["OnlineSecurity","OnlineBackup",'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','TotalCharges']
for i in g:
  sd2[i] = pd.to_numeric(sd[i],errors='coerce')
```

```python
print(sd2.shape)
print(sd2.head(1))
```

```python
print(sd2.info())
sd2=sd2.dropna()
```

```python
x=sd2.drop('Churn',axis=1)
y=sd2['Churn']
```

```python
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Assuming x and y are your feature and target variables
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
print(y_test.value_counts())
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
linear_model = LinearRegression()
ridge_model = Ridge()
lasso_model = Lasso()
random_forest_model = RandomForestRegressor()
```

```python
models=[linear_model,ridge_model,lasso_model,random_forest_model]
best_model=None
best_mae=float('inf')
for model in models:
  model.fit(x_train,y_train)
  y_pre=model.predict(x_test)
  mae=mean_squared_error(y_test,y_pre)
  print(f"current model is {type(model).__name__} current_mae is {mae}")
  if mae<best_mae:
    best_mae=mae
    best_model=model
print(f"best_model is {type(best_model).__name__},best mae is {best_mae}")
```
