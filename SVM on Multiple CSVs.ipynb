```python
# import zipfile
import shutil
import zipfile
# Path to the zip file you downloaded
zip_file_path = "/content/FIFA-20-complete-player-dataset.zip"

# Directory to extract the contents to in the local Colab environment
extract_dir = "/content/FIFA-20-complete-player-dataset"

# Create the directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Copy the zip file from Google Drive to local Colab environment
shutil.copy(zip_file_path, extract_dir)

# Unzip the file
zip_file_name = os.path.join(extract_dir, os.path.basename(zip_file_path))
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# List the contents of the extracted directory
extracted_files = os.listdir(extract_dir)
print("Extracted files:", extracted_files)
```

# **Using LabelEncoder for converting columns of categorical data in numeric data.**

```python
from sklearn.preprocessing import LabelEncoder
df4 = df3.select_dtypes(include=['object','category']).astype(str)
le=LabelEncoder()
df5=pd.DataFrame()
for i in df4.columns:
    label = le.fit_transform(df4[i])
    df5[i] = label
print(df5)
```

```python
print(df3.select_dtypes(include=['object']).columns)
```
